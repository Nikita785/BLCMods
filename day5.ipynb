```python

import pandas as pd
import numpy as np

def calculate_renewal_rates_from_df(df_financial, df_prolongations=None):
    """
    Рассчитывает коэффициенты пролонгации для каждого менеджера и для всего отдела в целом
    за каждый месяц и за год.

    Args:
        df_financial (pd.DataFrame): DataFrame с финансовыми данными, содержащий столбцы:
            - 'id': id проекта
            - Месяцы в формате 'Month Name YYYY': сумма отгрузки проекта в данный месяц (например, 'January 2023')
            - 'Account': ФИО ответственного аккаунт-менеджера
            - 'Month': дата последнего месяца реализации проекта (в формате 'YYYY-MM-DD')

        df_prolongations (pd.DataFrame, optional): DataFrame с данными о пролонгациях.  Если None, будет расчитано только по df_financial.
            Должен содержать столбцы:
            - 'id': id проекта
            - 'month': последний месяц реализации проекта
            - 'AM': ФИО ответственного аккаунт-менеджера (данные первичны по отношению к df_financial)
            Defaults to None.

    Returns:
        tuple:  (monthly_renewal_rates, annual_renewal_rates)
            monthly_renewal_rates (dict): Словарь, где ключи - названия месяцев, а значения - DataFrame с коэффициентами пролонгации для каждого менеджера и для всего отдела.
            annual_renewal_rates (pd.DataFrame): DataFrame с годовыми коэффициентами пролонгации для каждого менеджера и для всего отдела.
    """

    # Предобработка данных
    df = df_financial.copy()

    # 1. Преобразование столбца 'Month' в datetime
    df['Month'] = pd.to_datetime(df['Month'], errors='coerce') #errors='coerce' чтобы NaT обрабатывать

    # 2. Проверка и исправление типов данных
    month_cols = [col for col in df.columns if isinstance(col, str) and any(month in col.lower() for month in ['january', 'february', 'march', 'april', 'may', 'june', 'july', 'august', 'september', 'october', 'november', 'december'])]
    for col in month_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)  # Преобразуем в числовой тип, некорректные значения заменяем на 0

    # Замена NaT на None
    df = df.replace({pd.NaT: None})

    # 3. Фильтрация данных за 2023 год
    df = df[df['Month'].dt.year == 2023]

    # 4. Агрегируем данные по id и месяцу, суммируем значения, чтобы убрать дубликаты.
    df = df.groupby(['id', 'Account', 'Month'])[month_cols].sum().reset_index()

    # Если df_prolongations передан, то объединяем DataFrame.
    if df_prolongations is not None:
        # Проверка типов данных
        df_prolongations['month'] = pd.to_datetime(df_prolongations['month'])

        # Переименовываем столбец 'month' в 'prolongation_month' для ясности
        df_prolongations = df_prolongations.rename(columns={'month': 'prolongation_month'})

        # Объединяем df_financial и df_prolongations
        df = pd.merge(df, df_prolongations, left_on=['id', 'Account'], right_on=['id', 'AM'], how='left')

        # Удаляем дублирующийся столбец Account, если он есть.
        if 'Account_y' in df.columns:
            df = df.drop(columns=['Account_y'])
        df = df.rename(columns={'Account_x': 'Account'})

        df['AM'] = df['AM'].fillna(df['Account'])  # Заполняем пропуски в AM данными из Account

        # Фильтруем данные за 2023 год
        df = df[df['prolongation_month'].dt.year == 2023]

    else:
        df['AM'] = df['Account']
        df['prolongation_month'] = df['Month'] #Если нет df_prolongations то month берем из финансовых данных

    # Функция для расчета коэффициентов пролонгации
    def calculate_renewal_rates(df, target_month):
        """
        Рассчитывает коэффициенты пролонгации для заданного месяца.

        Args:
            df (pd.DataFrame): Объединенный DataFrame с данными о пролонгациях и финансах.
            target_month (pd.Timestamp): Месяц, для которого нужно рассчитать коэффициенты пролонгации.

        Returns:
            pd.DataFrame: DataFrame с коэффициентами пролонгации для каждого менеджера и для всего отдела.
        """

        # Рассчитываем начало и конец предыдущего и позапрошлого месяцев
        last_month_end = target_month - pd.DateOffset(months=1)
        last_month_start = last_month_end - pd.DateOffset(days=last_month_end.day - 1) # Первый день прошлого месяца

        previous_month_end = target_month - pd.DateOffset(months=2)
        previous_month_start = previous_month_end - pd.DateOffset(days=previous_month_end.day - 1)


        # Формируем строку месяца для поиска столбца.
        month_str = target_month.strftime('%B %Y')

        # Фильтруем данные для проектов, завершившихся в прошлом месяце
        projects_ended_last_month = df[(df['Month'] >= last_month_start) & (df['Month'] <= last_month_end)].copy()

        # Фильтруем данные для проектов, пролонгированных в первый месяц (целевой месяц)
        projects_renewed_first_month = df[(df['Month'] >= last_month_start) & (df['Month'] <= last_month_end) & (df['prolongation_month'] == target_month)].copy()

        # Фильтруем данные для проектов, завершившихся в позапрошлом месяце и НЕ пролонгированных в прошлом месяце
        projects_not_renewed_first_month = df[(df['Month'] >= previous_month_start) & (df['Month'] <= previous_month_end)].copy()
        projects_not_renewed_first_month = projects_not_renewed_first_month[~projects_not_renewed_first_month['id'].isin(projects_ended_last_month['id'].unique())]

        # Фильтруем данные для проектов, пролонгированных во второй месяц (целевой месяц)
        projects_renewed_second_month = df[(df['Month'] >= previous_month_start) & (df['Month'] <= previous_month_end) & (df['prolongation_month'] == target_month) & (df['id'].isin(projects_not_renewed_first_month['id'].unique()))].copy()

        # Функция для расчета коэффициента пролонгации
        def calculate_rate(group, ended_projects, renewed_projects, month_str):
            total_volume = ended_projects[month_str].sum() if not ended_projects.empty and month_str in ended_projects.columns else 0
            renewed_volume = renewed_projects[month_str].sum() if not renewed_projects.empty and month_str in renewed_projects.columns else 0
            rate = renewed_volume / total_volume if total_volume > 0 else 0
            return rate

        # Расчет для первого месяца
        def calculate_rate_first_month(group, month_str):
            ended_projects = projects_ended_last_month[projects_ended_last_month['AM'] == group.name]
            renewed_projects = projects_renewed_first_month[projects_renewed_first_month['AM'] == group.name]
            return calculate_rate(group, ended_projects, renewed_projects, month_str)

        # Расчет для второго месяца
        def calculate_rate_second_month(group, month_str):
            ended_projects = projects_not_renewed_first_month[projects_not_renewed_first_month['AM'] == group.name]
            renewed_projects = projects_renewed_second_month[projects_renewed_second_month['AM'] == group.name]
            return calculate_rate(group, ended_projects, renewed_projects, month_str)

        # Расчет по менеджерам для первого месяца
        renewal_rates_first_month = df.groupby('AM').apply(lambda x: calculate_rate_first_month(x, month_str)).reset_index(name='Renewal_Rate_First_Month')

        # Расчет по менеджерам для второго месяца
        renewal_rates_second_month = df.groupby('AM').apply(lambda x: calculate_rate_second_month(x, month_str)).reset_index(name='Renewal_Rate_Second_Month')

        # Расчет для всего отдела (первый месяц)
        total_ended_volume = projects_ended_last_month[month_str].sum() if not projects_ended_last_month.empty and month_str in projects_ended_last_month.columns else 0
        total_renewed_volume = projects_renewed_first_month[month_str].sum() if not projects_renewed_first_month.empty and month_str in projects_renewed_first_month.columns else 0
        overall_renewal_rate_first_month = total_renewed_volume / total_ended_volume if total_ended_volume > 0 else 0

        # Расчет для всего отдела (второй месяц)
        total_ended_volume_2 = projects_not_renewed_first_month[month_str].sum() if not projects_not_renewed_first_month.empty and month_str in projects_not_renewed_first_month.columns else 0
        total_renewed_volume_2 = projects_renewed_second_month[month_str].sum() if not projects_renewed_second_month.empty and month_str in projects_renewed_second_month.columns else 0
        overall_renewal_rate_second_month = total_renewed_volume_2 / total_ended_volume_2 if total_ended_volume_2 > 0 else 0

        # Объединяем результаты
        renewal_rates = pd.merge(renewal_rates_first_month, renewal_rates_second_month, on='AM', how='outer')

        # Добавляем строку с общим результатом по отделу
        renewal_rates = renewal_rates.append({'AM': 'Overall', 'Renewal_Rate_First_Month': overall_renewal_rate_first_month, 'Renewal_Rate_Second_Month': overall_renewal_rate_second_month}, ignore_index=True)

        return renewal_rates


    # Список месяцев для анализа (все месяцы 2023 года)
    months_2023 = pd.date_range(start='2023-01-01', end='2023-12-01', freq='MS')

    # Рассчитываем коэффициенты для каждого месяца
    monthly_renewal_rates = {}
    for month in months_2023:
        monthly_renewal_rates[month.strftime('%B')] = calculate_renewal_rates(df.copy(), month)


    # Годовой расчет (агрегируем месячные данные)
    def calculate_annual_renewal_rates(monthly_renewal_rates):
        """
        Рассчитывает годовые коэффициенты пролонгации на основе месячных данных.

        Args:
            monthly_renewal_rates (dict): Словарь, где ключи - названия месяцев, а значения - DataFrame с месячными коэффициентами.

        Returns:
            pd.DataFrame: DataFrame с годовыми коэффициентами пролонгации для каждого менеджера и для всего отдела.
        """
        annual_data = {}
        for month, rates in monthly_renewal_rates.items():
            for _, row in rates.iterrows():
                am = row['AM']
                if am not in annual_data:
                    annual_data[am] = {'AM': am, 'Total_First_Month_Volume': 0, 'Total_First_Month_Renewed': 0, 'Total_Second_Month_Volume': 0, 'Total_Second_Month_Renewed': 0}

                # Получаем данные для расчета
                target_month = pd.to_datetime(month, format='%B').replace(year=2023)
                month_str = target_month.strftime('%B %Y')

                # Рассчитываем начало и конец предыдущего и позапрошлого месяцев
                last_month_end = target_month - pd.DateOffset(months=1)
                last_month_start = last_month_end - pd.DateOffset(days=last_month_end.day - 1) # Первый день прошлого месяца

                previous_month_end = target_month - pd.DateOffset(months=2)
                previous_month_start = previous_month_end - pd.DateOffset(days=previous_month_end.day - 1)

                # Фильтруем данные для проектов, завершившихся в прошлом месяце
                projects_ended_last_month = df[(df['Month'] >= last_month_start) & (df['Month'] <= last_month_end) & (df['AM'] == am)].copy()

                # Фильтруем данные для проектов, пролонгированных в первый месяц (целевой месяц)
                projects_renewed_first_month = df[(df['Month'] >= last_month_start) & (df['Month'] <= last_month_end) & (df['prolongation_month'] == target_month) & (df['AM'] == am)].copy()

                # Фильтруем данные для проектов, завершившихся в позапрошлом месяце и НЕ пролонгированных в прошлом месяце
                projects_not_renewed_first_month = df[(df['Month'] >= previous_month_start) & (df['Month'] <= previous_month_end) & (df['AM'] == am)].copy()
                projects_not_renewed_first_month = projects_not_renewed_first_month[~projects_not_renewed_first_month['id'].isin(projects_ended_last_month['id'].unique())]

                # Фильтруем данные для проектов, пролонгированных во второй месяц (целевой месяц)
                projects_renewed_second_month = df[(df['Month'] >= previous_month_start) & (df['Month'] <= previous_month_end) & (df['prolongation_month'] == target_month) & (df['id'].isin(projects_not_renewed_first_month['id'].unique()))  & (df['AM'] == am)].copy()


                annual_data[am]['Total_First_Month_Volume'] += projects_ended_last_month[month_str].sum() if not projects_ended_last_month.empty and month_str in projects_ended_last_month.columns else 0
                annual_data[am]['Total_First_Month_Renewed'] += projects_renewed_first_month[month_str].sum() if not projects_renewed_first_month.empty and month_str in projects_renewed_first_month.columns else 0
                annual_data[am]['Total_Second_Month_Volume'] += projects_not_renewed_first_month[month_str].sum() if not projects_not_renewed_first_month.empty and month_str in projects_not_renewed_first_month.columns else 0
                annual_data[am]['Total_Second_Month_Renewed'] += projects_renewed_second_month[month_str].sum() if not projects_renewed_second_month.empty and month_str in projects_renewed_second_month.columns else 0

        # Преобразуем в DataFrame
        annual_df = pd.DataFrame.from_dict(annual_data, orient='index')

        # Рассчитываем коэффициенты
        annual_df['Annual_Renewal_Rate_First_Month'] = annual_df['Total_First_Month_Renewed'] / annual_df['Total_First_Month_Volume']
        annual_df['Annual_Renewal_Rate_Second_Month'] = annual_df['Total_Second_Month_Renewed'] / annual_df['Total_Second_Month_Volume']

        # Выбираем нужные столбцы
        annual_df = annual_df[['AM', 'Annual_Renewal_Rate_First_Month', 'Annual_Renewal_Rate_Second_Month']]

        return annual_df

    # Рассчитываем годовые коэффициенты
    annual_renewal_rates = calculate_annual_renewal_rates(monthly_renewal_rates)

    return monthly_renewal_rates, annual_renewal_rates

# Пример использования:
# (monthly_renewal_rates, annual_renewal_rates) = calculate_renewal_rates_from_df(df_financial, df_prolongations)
```
