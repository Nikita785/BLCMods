```python
import pandas as pd

df_fd = pd.read_csv('/content/financial_data.csv')
df_pr = pd.read_csv('/content/prolongations.csv')
```

```python
#  Шаг первый: очистка данных

  # df_fd

#    - Удалил столбец "Причины дубля" из df_fd
df_fd = df_fd.drop('Причина дубля', axis=1)

#    - Преобразовал месяцы в df_fd в числовой тип и заполнил NaN нулями
df_fd_dt = ['id', 'Account']

for col in df_fd.columns:
  if col not in df_fd_dt:
    df_fd[col] = pd.to_numeric(df_fd[col], errors='coerce').fillna(0)

#     - Преобразовал названия столбцов с месяцами на английский

month_translation = {
    'Январь': 'January',
    'Февраль': 'February',
    'Март': 'March',
    'Апрель': 'April',
    'Май': 'May',
    'Июнь': 'June',
    'Июль': 'July',
    'Август': 'August',
    'Сентябрь': 'September',
    'Октябрь': 'October',
    'Ноябрь': 'November',
    'Декабрь': 'December'
}

for col in df_fd.columns:
    month_name_parts = col.split()
    if len(month_name_parts) > 1 and month_name_parts[0] in month_translation:
        new_col = month_translation[month_name_parts[0]] + ' ' + month_name_parts[1]
        df_fd.rename(columns={col: new_col}, inplace=True)

  # df_pr
month_translation_dr = {
    'январь': 'January',
    'февраль': 'February',
    'март': 'March',
    'апрель': 'April',
    'май': 'May',
    'июнь': 'June',
    'июль': 'July',
    'август': 'August',
    'сентябрь': 'September',
    'октябрь': 'October',
    'ноябрь': 'November',
    'декабрь': 'December'
}

#Переименовал столбцы
df_pr.rename(columns={'AM': 'Account', 'month': 'Month'}, inplace=True)

#Преобразовал значения столбца

for index, value in df_pr['Month'].items():
  value_parts = str(value).split()
  if len(value_parts) > 1 and value_parts[0].lower() in month_translation_dr:
    new_val = month_translation_dr[value_parts[0].lower()] + ' ' + value_parts[1]
    df_pr.loc[index, 'Month'] = new_val
```

```python
# Шаг второй: объединение таблиц для произведения расчётов


df_merge = pd.merge(df_fd, df_pr, on=['id', '', how='left')
```
